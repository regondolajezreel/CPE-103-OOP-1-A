Create a Python class fo Student

```python
class Student():
    def __init__(self, name, sex, weight, height, id, program):
        self.name = name
        self.sex = sex
        self.weight = weight
        self.height = height
        self.id = id
        self.program = program

    def description(self):
      return (f"Name: {self.name}\n"
              f"Sex: {self.sex}\n"
              f"Height: {self.height}\n"
              f"Weight: {self.weight}\n"
              f"ID: {self.id}\n"
              f"Program: {self.program}")
```

```python
class gradeCalcu():

  def __init__(self, name):
    self.name = name
    self.mQ, self.fQ = 0, 0
    self.mS, self.fS= 0, 0
    self.mA, self.fA = 0, 0
    self.mR, self.fR = 0, 0

  def grade(self, gradeType):
    print(f"\nEnter the grades for {gradeType}: ")
    while True:
      try:
        quiz = float(input(f"Enter {gradeType} Quiz average (out of 100): "))
        assignment = float(input(f"Enter {gradeType} Assignment average (out of 100): "))
        seatwork = float(input(f"Enter {gradeType} Seatwork average (out of 100): "))
        recitation = float(input(f"Enter {gradeType} Recitation average (out of 100): "))

        if 0 <= quiz <= 100 and 0 <= assignment <= 100 and 0 <= seatwork <= 100 and 0 <= recitation <= 100:
          break
        else:
          print("Please enter valid scores between 0 and 100.")
      except ValueError:
        print("Invalid input. Please enter a valid number.")

    if gradeType == "Midterm":
      self.mQ = quiz
      self.mS = assignment
      self.mA = seatwork
      self.mR = recitation
    elif gradeType == "Final":
      self.fQ = quiz
      self.fA = assignment
      self.fS = seatwork
      self.fR = recitation

  def calculateMidtermGrade(self): #Calculate Midterm Grade
        return (0.25 * self.mQ) + (0.25 * self.mS) + (0.25 * self.mA) + (0.25 * self.mR)

  def calculateFinalsGrade(self, midtermGrade): #Calculate Final Grade
        return (1 / 3) * midtermGrade + (2 / 3) * (self.fQ + self.fA + self.fS + self.fR) / 4

  def calculateGWA(self, midtermGrade, finalGrade):
    return (1/3 * midtermGrade) + (2/3 * finalGrade)
```
