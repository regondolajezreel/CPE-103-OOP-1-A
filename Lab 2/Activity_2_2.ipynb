```python
def main():
  class calculateGrade():
    def __init__(self, name):
      self.name = name
      self.PQ, self.MQ, self.FQ = 0, 0, 0
      self.PAS, self.MAS, self.FAS = 0, 0, 0
      self.PHOA, self.MHOA, self.FHOA = 0, 0, 0
      self.PE, self.ME, self.FE = 0, 0, 0

    def gradeInput(self, gradeType):

      print(f"\nEnter the grades for {gradeType}: ")
      while True:
        try:
          quiz = float(input(f"Enter {gradeType} Quiz average (out of 100): "))
          assignment = float(input(f"Enter {gradeType} Assignment average (out of 100): "))
          handsOn = float(input(f"Enter {gradeType} Hands-on activity average (out of 100): "))
          exam = float(input(f"Enter {gradeType} Exam score (out of 100): "))

          # Check if scores are within valid ranges
          if 0 <= quiz <= 100 and 0 <= assignment <= 100 and 0 <= handsOn <= 100 and 0 <= exam <= 100:
            break
          else:
            print("Please enter valid scores between 0 and 100.")
        except ValueError:
          print("Invalid input. Please enter a valid number.")

      if gradeType == "Prelim":
        self.PQ = quiz
        self.PAS = assignment
        self.PHOA = handsOn
        self.PE = exam
      elif gradeType == "Midterm":
        self.MQ = quiz
        self.MAS = assignment
        self.MHOA = handsOn
        self.ME = exam
      elif gradeType == "Final":
        self.FQ = quiz
        self.FAS = assignment
        self.FHOA = handsOn
        self.FE = exam

    def calculatePrelimCs(self):#Calculate Prelim Class Standing
          return (0.5 * self.PHOA) + (0.3 * self.PQ) + (0.2 * self.PAS)

    def calculatePrelimGrade(self): #Calculate Prelim Grade
          prelimCs = self.calculatePrelimCs()
          return (0.5 * self.PE) + (0.5 * prelimCs)

    def calculateMidtermCs(self): #Calculate Midterm Class Standing
          return (0.5 * self.MHOA) + (0.3 * self.MQ) + (0.2 * self.MAS)

    def calculateMidtermGrade(self, prelimGrade): #Calculate Midterm Grade
          midtermCs = self.calculateMidtermCs()
          return ((1 / 3) * prelimGrade) + ((2 / 3) * ((0.5 * self.ME) + (0.5 * midtermCs)))

    def calculateFinalsCs(self): #Calculate Final Class Standing
          return (0.5 * self.FHOA) + (0.3 * self.FQ) + (0.2 * self.FAS)

    def calculateFinalsGrade(self, midtermGrade): #Calculate Finals Grade
          finalsCs = self.calculateFinalsCs()
          return ((1 / 3) * midtermGrade) + ((2 / 3) * ((0.5 * self.FE) + (0.5 * finalsCs)))

    def calculateFinalGrade(self, midtermGrade, finalsGrade): #Calculate Final Grade
          return ((1/3) * midtermGrade) + ((2/3) * finalsGrade)



  #Ask for student's name
  studentName = input("Please enter the student's name: ")
  calculate = calculateGrade(studentName)

  # Input grades for Prelim, Midterm, and Final
  for gradeType in ["Prelim", "Midterm", "Final"]:
    calculate.gradeInput(gradeType)

    # Calculate Prelim Grade
    prelimGrade = calculate.calculatePrelimGrade()

    # Calculate Midterm Grade
    midtermGrade = calculate.calculateMidtermGrade(prelimGrade)

    # Calculate Finals Grade
    finalsGrade = calculate.calculateFinalsGrade(midtermGrade)

    #Calculate Final Grade
    finalGrade = calculate.calculateFinalGrade(midtermGrade, finalsGrade)

  # Display the final results
  print(f"\nName: {calculate.name} ")
  print(f"Prelim Grade: {prelimGrade:.2f}")
  print(f"Midterm Grade: {midtermGrade:.2f}")
  print(f"Finals Grade: {finalsGrade:.2f}")
  return finalGrade

main()
```
